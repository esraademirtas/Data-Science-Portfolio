```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("12-health_risk_classification.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#scale edilmesine Ã§ok gerek yok temiz denebilecek bi veri 
```

```python
sns.scatterplot(x=df["blood_pressure_variation"], y=df["activity_level_index"], hue=df["high_risk_flag"])
plt.show()
```

```python
df["high_risk_flag"].value_counts()
```

```python
X = df.drop("high_risk_flag", axis=1)
y = df["high_risk_flag"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
sns.boxenplot(df)
plt.show()
```

```python
#scaler kullanmasak da olur ama yine de yapalÄ±m 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#neighbors parametresi en Ã¶nemlisi 
classifier = KNeighborsClassifier(n_neighbors=5, algorithm="auto", weights="uniform")
```

```python
classifier.fit(X_train_scaled, y_train)
```

```python
y_pred = classifier.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print("confusion matrics: \n", confusion_matrix(y_pred, y_test))
print("accuracy score: ", accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
y_pred = classifier.predict(X_test_scaled)
```

```python
y_pred = classifier.predict(X_test_scaled)
```

```python
print("confusion matrics: \n", confusion_matrix(y_pred, y_test))
print("accuracy score: ", accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))  #baya dÃ¼ÅtÃ¼ en yakÄ±n 70 komÅuya baktÄ±ÄÄ± iÃ§in
```

```python
X = df_reg.drop("daily_energy_consumption_kwh", axis=1)
y = df_reg["daily_energy_consumption_kwh"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
regressor = KNeighborsRegressor(n_neighbors=5, algorithm="auto")
```

```python
regressor.fit(X_train_scaled, y_train)
```

```python
y_pred = regressor.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print("r2 score: ", r2_score(y_pred, y_test))
print("mean absolute error: ", mean_absolute_error(y_pred, y_test))
print("mean squared error: ", mean_squared_error(y_pred, y_test))
```

```python
plt.scatter(y_pred, y_test)
```

```python
#parametrelerle oynayalÄ±m
```

```python
regressor = KNeighborsRegressor(n_neighbors=7, algorithm="auto")
regressor.fit(X_train_scaled, y_train)
y_pred = regressor.predict(X_test_scaled)
print("r2 score: ", r2_score(y_pred, y_test))
print("mean absolute error: ", mean_absolute_error(y_pred, y_test))
print("mean squared error: ", mean_squared_error(y_pred, y_test))
```

```python
regressor = KNeighborsRegressor(n_neighbors=30, algorithm="auto")
regressor.fit(X_train_scaled, y_train)
y_pred = regressor.predict(X_test_scaled)
print("r2 score: ", r2_score(y_pred, y_test))
print("mean absolute error: ", mean_absolute_error(y_pred, y_test))
print("mean squared error: ", mean_squared_error(y_pred, y_test))
```
